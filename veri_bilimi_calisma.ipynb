## Machine Learning

```python
df.head()
```

```python
island = pd.get_dummies(df['island'], drop_first=True)
```

```python
df2 = pd.concat([df, island], axis=1).drop(['island'], axis=1)
```

```python
df2.head()
```

```python
target, features = df2.species, df2.drop('species', axis=1)
```

```python
features.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
features.iloc[:,:4]
```

```python
scaler.fit(features.iloc[:,:4])
```

```python
features.iloc[:,:4] = scaler.transform(features.iloc[:,:4])
```

```python
features
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
target_encoded = le.fit_transform(target)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, target_encoded, test_size=0.2, random_state=41)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
```

```python
preds = tree.predict(X_test)
```

```python
confusion_matrix(y_test, preds)
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(confusion_matrix(y_test, preds), annot=True, cbar=False)
```

```python
print(classification_report(y_test, preds))
```

```python
print('Accuracy Score= %', accuracy_score(y_test, preds) *100)
```

```python
final_df = pd.concat([features, pd.Series(target_encoded, name='Target')], axis=1)
```

```python
final_df
```

```python
final_df.to_csv('penguins_scaled.csv', index=False, header=False)
```
